##  Handling Missing / Incorrect Values


```python
print("\nMissing Values Before Cleaning:\n")
print(df.isnull().sum())
```

```python
# Fix TotalCharges column (contains blanks as strings)
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
# Fill missing values with 0 (or you can use mean/median)
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
print("\nMissing Values After Cleaning:\n")
print(df.isnull().sum())
```

```python
# Feature Engineering
```

```python
# Convert SeniorCitizen to categorical
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
```

```python
# Encoding Categorical Variables
```

```python
# Identify categorical columns (excluding target + ID)
categorical_cols = [
    col for col in df.columns 
    if df[col].dtype == 'object' and col not in ['customerID', 'Churn']
]
```

```python
# One-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
# Convert target variable
df_encoded['Churn'] = df_encoded['Churn'].map({'No': 0, 'Yes': 1})
```
